```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fc = pd.read_csv("FuelConsumptionCo2.csv")
```

```python
fc.head()
```

```python
fc.info()
```

```python
# To know the brands existing  

fc["Brands"].unique()
```

```python
# To know thy type of fuel used by each brrand

fc.groupby("Brands")["FUELTYPE"].value_counts()
```

```python
# To know the number of cylenders in each car class

fc.groupby("VEHICLECLASS")["CYLINDERS"].value_counts()
```

```python
# To know the distribution of the value of the co2 cons.

sns.distplot(fc["CO2EMISSIONS"])
```

```python
# To know the co2 emission ammount accordind to the fuel type

sns.barplot(x="FUELTYPE" , y="CO2EMISSIONS" , data=fc)
```

```python
# To know the Fuel consumption accordind to the fuel type

sns.barplot(x="FUELTYPE" , y="FUELCONSUMPTION_COMB_MPG" , data=fc)
```

```python
# To show the relation betwee co2 emmission and fuel con.

sns.jointplot(x="CO2EMISSIONS", y="FUELCONSUMPTION_COMB_MPG" , data=fc, kind="scatter")
```
